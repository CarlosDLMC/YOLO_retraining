```python
from ultralytics import YOLO
from PIL import Image

PROJECT_DIR = "/home/mentefria/Documents/YOLO_retraining"

# model = YOLO(f"{PROJECT_DIR}/models/best_100_ep_dtv_4.pt")
model = YOLO(f"{PROJECT_DIR}/runs/detect/train13/weights/best.pt")
```

```python
def print_predictions_for_image(model, image_path):
    results = model(image_path)
    im_array = results[0].plot()
    im = Image.fromarray(im_array[..., ::-1])
    display(im)
```

```python
print_predictions_for_image(model, f"{PROJECT_DIR}/images/test_images/teslas.jpg")
```
